```python
import automol

re_smi = ("CC(O[O])CC",)
pr_smi = ("CC=CC", "O[O]")

rxns = automol.reac.from_smiles(re_smi, pr_smi, stereo=False)
```

```python
(rxn,) = rxns
```

### Expanding the stereoisomers

The best thing to do for a database is only store the canonical amchi (i.e. m0) and only run the calculations on canonical amchis

```python
for stereo_rxn in automol.reac.expand_stereo(rxn, enant=False):
    # Enant = false disables duplicate enantiomers
    amchi = automol.reac.ts_amchi(stereo_rxn)
    print(amchi)

    reactant_amchis, product_amchis = automol.reac.amchis(stereo_rxn)
    print(reactant_amchis)
    print(product_amchis)

    # automol.graph.display(stereo_rxn)
    geom = automol.graph.geometry(stereo_rxn)
    print(type(geom))
    automol.geom.display(geom)
    # xyz = automol.geom.xyz_string(geom)
```

```python
dct = {}

for i, rxn in enumerate(rxns):
    automol.reac.display(rxn)
    dct[i] = automol.reac.ts_amchi(rxn)
```
